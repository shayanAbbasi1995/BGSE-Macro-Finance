```python
import pandas as pd
import numpy as np
```

```python
final_lit = pd.read_csv('final_lit.csv')
firm_code = pd.read_stata('Firm-code-info.dta')
```

```python
for i in range(len(final_lit.final_conm)):
  if pd.notna(final_lit.final_conm.iloc[i]):
    final_lit.final_conm.iloc[i] = final_lit.final_conm.iloc[i][1:-1] 
```

```python
for i in range(len(final_lit['Parent Company Stock Ticker'])):
  if pd.notna(final_lit['Parent Company Stock Ticker'].iloc[i]):
    final_lit['Parent Company Stock Ticker'].iloc[i] = final_lit['Parent Company Stock Ticker'].iloc[i].replace('LSE: ', '')
    final_lit['Parent Company Stock Ticker'].iloc[i] = final_lit['Parent Company Stock Ticker'].iloc[i].replace('NYSE: ', '')
    final_lit['Parent Company Stock Ticker'].iloc[i] = final_lit['Parent Company Stock Ticker'].iloc[i].replace('Nasdaq: ', '')
```

```python
firm_code['conm'] = firm_code['conm'].str.lower()
```

```python
final_lit
```

```python
new_df = pd.merge(final_lit, firm_code,  how='left', left_on=['Penalty Year','final_conm'], right_on = ['fyear','conm'])
```

```python
new_df2 = pd.merge(final_lit, firm_code,  how='left', left_on=['Penalty Year','Parent Company Stock Ticker'], right_on = ['fyear','tic'])
```

```python
new_df.update(new_df2)
```
